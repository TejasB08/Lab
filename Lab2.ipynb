```python
import tensorflow as tf
import unittest

# function with nested operations
def compute_nested_graph(x, y):
    a= tf.multiply(x, y)
    b= tf.add(a, x)
    c= tf.square(b)
    return c


#Unit test using tf.test.Testcase
class TestTensorflowGraph(tf.test.TestCase):
    def test_nested_operations(self):
        x = tf.constant(3.0)
        y = tf.constant(2.0)
        result = compute_nested_graph(x, y)
        self.assertEqual(self.evaluate(result), (3*2+3)**2)


#Run test
unittest.main(argv=[''], verbosity=2, exit=False)
```
